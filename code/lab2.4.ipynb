```python
from thinkdsp import TriangleSignal

# Создаем треугольный сигнал с частотой 440 Гц
triangle_signal = TriangleSignal(440)

# Создаем волну длительностью 0.01 секунды
duration = 0.01
wave = triangle_signal.make_wave(duration=duration)
# Выводим сигнал на экран
wave.plot()
```

```python
# Создаем объект Spectrum
spectrum = wave.make_spectrum()

# Распечатываем значение Spectrum.hs[0]
print("Амплитуда компоненты с частотой 0:", spectrum.hs[0])
```

```python
# Устанавливаем значение Spectrum.hs[0] = 100
spectrum.hs[0] = 100

# Создаем новую волну, соответствующую измененному спектру
new_wave = spectrum.make_wave()

# Выводим новую волну на экран
new_wave.plot()
```

Видно, что сигнал теперь полностью положительный
