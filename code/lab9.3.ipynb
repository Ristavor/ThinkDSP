```python
import os

import numpy as np
import matplotlib.pyplot as plt

from thinkdsp import decorate

from thinkdsp import TriangleSignal
```

```python
# прямоугольный сигнал
from thinkdsp import SquareSignal

in_wave = SquareSignal(freq=50).make_wave(duration=0.1, framerate=44100)
in_wave.plot()
decorate(xlabel='Time (s)')
```

```python
out_wave = in_wave.cumsum()
out_wave.plot()
decorate(xlabel='Time (s)')
```

Результатом применения cumsum является треугольный сигнал

```python
spectrum = in_wave.make_spectrum().integrate()
spectrum.hs[0] = 0
out_wave2 = spectrum.make_wave()
out_wave2.plot()
decorate(xlabel='Time (s)')
```

Результатом применения integrate является треугольный сигнал

```python
out_wave.unbias()
out_wave.normalize()
out_wave2.normalize()
out_wave.plot()
out_wave2.plot()
```

```python
out_wave.max_diff(out_wave2)
```

Результаты преобразования обоих волн почти идентичные, но есть небольшая разницы.
