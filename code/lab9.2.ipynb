```python
import os

import numpy as np
import matplotlib.pyplot as plt

from thinkdsp import decorate
# создадим треугольный сигнал
from thinkdsp import TriangleSignal

in_wave = TriangleSignal(freq=50).make_wave(duration=0.1, framerate=44100)
in_wave.plot()
decorate(xlabel='Time (s)')
```

```python
out_wave = in_wave.diff()
out_wave.plot()
decorate(xlabel='Time (s)')
#результат diff
```

```python
out_wave2 = in_wave.make_spectrum().differentiate().make_wave()
out_wave2.plot()
decorate(xlabel='Time (s)')
# результат differtiate
```

В местах изменения наклона треугольного сигнала в differntiate присутствует "звон" ввиду неопределенности в точках треугольника
