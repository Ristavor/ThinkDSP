```python
import matplotlib.pyplot as plt
from thinkdsp import SquareSignal, decorate
```

```python
def modify_spectrum(spectrum):
    """Modify the spectrum by dividing each element of hs by its corresponding frequency from fs."""
    # Set Spectrum.hs[0] to 0 to avoid division by zero
    spectrum.hs[0] = 0
    # Divide each element of hs by its corresponding frequency from fs
    spectrum.hs[1:] /= spectrum.fs[1:]
    return spectrum
```

```python
# Создаем прямоугольный сигнал с частотой 200 Гц
square_signal = SquareSignal(200)

# Создаем волну
duration = 5
wave = square_signal.make_wave(duration, framerate=10000)

# Вычисляем спектр и распечатываем его
spectrum = wave.make_spectrum()
spectrum.plot()
decorate(xlabel='Frequency (Hz)')
plt.title("Spectrum before modification")
plt.show()

# Изменяем спектр
spectrum_new = modify_spectrum(spectrum)

# Печатаем спектр после изменений
spectrum_new.plot()
decorate(xlabel='Frequency (Hz)')
plt.title("Spectrum after modification")
plt.show()

```

```python
wave.make_audio()
```

```python
wave2 = spectrum_new.make_wave()
wave2.make_audio()
```
