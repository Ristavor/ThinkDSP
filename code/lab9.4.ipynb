```python
#Создадим пилообразный сигнал
from thinkdsp import SawtoothSignal, decorate

in_wave = SawtoothSignal(freq=50).make_wave(duration=0.1, framerate=44100)
in_wave.plot()
decorate(xlabel='Time (s)')
```

```python
#Применим дважды cumsum
out_wave = in_wave.cumsum()
out_wave.unbias()
out_wave.plot()
decorate(xlabel='Time (s)')
```

```python
out_wave = out_wave.cumsum()
out_wave.plot()
decorate(xlabel='Time (s)')
```

```python
#Теперь применим дважды интегрирование
spectrum = in_wave.make_spectrum().integrate().integrate()
spectrum.hs[0] = 0
out_wave2 = spectrum.make_wave()
out_wave2.plot()
decorate(xlabel='Time (s)')
```

```python
#На данный момент результат все больше и больше напоминает синусоиду. Причина в том, что интегрирование действует как фильтр нижних частот. На данный момент мы отфильтровали почти все, кроме фундаментального, как показано в приведенном ниже спектре:
out_wave2.make_spectrum().plot(high=500)
```
