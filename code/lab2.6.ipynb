```python
from thinkdsp import SawtoothSignal
import numpy as np

#   
sawtooth_signal = SawtoothSignal()

sawtooth_signal.make_wave(duration=0.1, framerate=10000).plot()

```

```python
#   
spectrum = sawtooth_signal.make_wave(duration=0.1, framerate=10000).make_spectrum()

#  :      1/f^2
spectrum.hs[0] = 0
spectrum.hs[1:] /= spectrum.fs[1:]

#          
wave = spectrum.make_wave()
wave.plot()
```

```python
wave.segment(duration = 0.005).plot()
```
