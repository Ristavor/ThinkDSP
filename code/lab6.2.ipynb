```python
from thinkdsp import read_wave

wave = read_wave('328878__tzurkan__guitar-phrase-tzu.wav')
wave.plot()
wave.make_audio()
```

Выберем какой-нибудь сегмент

```python
segment = wave.segment(start = 1,duration = 1.0)
segment.normalize()
segment.make_audio()
```

```python
dct = segment.make_dct()
dct.plot(high = 5000)
```

```python
def compress(dct,limit = 0):
  for i, amp in enumerate(dct.amps):
    if np.abs(amp) < limit:
      dct.hs[i] = 0
```

```python
compress(dct,1000)
```

```python
dct.plot(high = 5000)
```

```python
dct.make_wave().make_audio()
```

При 1000 разница заметная. Попробуем уменьшить

```python
dct = segment.make_dct()
compress(dct,100)
dct.make_wave().make_audio()
```

При 100 результат довольно похожий

```python
dct = segment.make_dct()
compress(dct,400)
dct.make_wave().make_audio()
```
